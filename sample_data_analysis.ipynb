# Data analysis with Python.

I am using Google Colaboratory and the Pandas Python library to analyze sample data from a Far Horizons mission to the edge of space. 

In this guide, I will take the sample data from the spreedsheet and organize it into tables.

## Import Pandas Python library

```
import pandas as pd
```

## Import sensor data file

```
from google.colab import files

uploaded = files.upload()
```

## Read sensor data file

```
df = pd.read_csv("sampledata.csv")
```

## Explore the data

Look at the columns

```
#Tell me the information that is in the first column of the spreadsheet.
#In this case, df means data frame.
df.columns
```

Select only the columns with names

```
#Pick out only the columns with names for the table we are about to create. 
#Data frame 1 of our table will include these named columns from the spreadsheet:
df1 = df[['Launch+','Launch+.1','time','lat','long','alt','Temp','pressure','accel','accel.1','accel.2','p','T(p)','RH','T(RH)','T','Geiger','Geiger.1','Geiger.2']]
```

How many observations are there?

```
#In this case, len stands for length. We are looking for how many rows of data the data frame will have.
len(df1)
```

Look at the first 20 Rows, sorted by Time. 

```
df1.sort_values(by="time",ascending=True).head(20)
```

Look at the last 20 rows, sorted by time. 

```
df1.sort_values(by="time",ascending=True).tail(20)
```
